```python
import tensorflow as tf
```

### 사용데이터

```python
train_x = [1, 2, 3, 4, 5, 6, 7]
train_y = [3, 5, 7, 9, 11, 13, 15]
```

### 모델만들기

#### pred_y = train_x * a + b의 형식으로 예측

```python
a = tf.Variable(0.1) # 변수초기값은 randomize
b = tf.Variable(0.1)
```

### 손실함수

```python
def lossFunction(): # mse 구현
    pred_y = train_x * a + b
    return tf.keras.losses.mse(train_y, pred_y)
```

### 경사하강법

```python
opt = tf.keras.optimizers.Adam(learning_rate = 0.001)
```

```python
for i in range(20000):
    opt.minimize(lossFunction, var_list=[a, b])
    # 손실함수에 lambda:손실함수(a, b) 형식으로 사용
    print(a.numpy(),b.numpy())
```
