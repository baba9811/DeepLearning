키와 발사이즈를 선형모델로 추정

```python
height = 170
feet = 260
```

```python
a = tf.Variable(0.1)
b = tf.Variable(0.2)
```

```python
def lossFunction():
    pred = height * a + b
    return tf.square(260- pred) # (실제값 - 예측값)^2, tf.square(실제값 - 예측값)
```

```python
opt = tf.keras.optimizers.Adam(learning_rate = 0.1) # 경사하강법 module
```

```python
opt.minimize(lossFunction, var_list=[a, b]) # 경사하강 1번
```

```python
for i in range(300):
    opt.minimize(lossFunction, var_list = [a, b])
    print(a.numpy(), b.numpy())
```
