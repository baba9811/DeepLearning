```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

### 데이터로드 및 전처리

#### 1. pandas로 데이터 로드

```python
data = pd.read_csv("data/gpascore.csv")
```

```python
data.head()
```

admit: 합격여부, gre: 영어성적, gpa: 학점, rank: 지원한 대학교 레벨

#### 2. 전처리

```python
data.info()
```

```python
print(data.isna().sum())
```

gre column 에 하나의 결측치 존재

```python
data.dropna(inplace = True) # 결측치를 제거
```

```python
data.reset_index(inplace=True, drop=True)
```

data.fillna()를 사용하면 임의의 값으로 대체

```python
data.describe()
```

#### 3. x,y 데이터 구분

```python
y = data['admit']
```

```python
x = data.iloc[:, 1:]
```

```python
x.head()
```

```python
y.head()
```

#### 다음과 같은 방식으로 x,y 데이터를 배열로 만들 수 있음

```python
y_train = data['admit'].values
```

```python
y_train
```

```python
x_train = []
for i, rows in data.iterrows(): # iterrows()는 데이터프레임의 한 행씩 출력
    x_train.append([rows['gre'], rows['gpa'], rows['rank']])
```

```python
x_train
```

### 딥러닝 모델 디자인

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu'), # hidden layer
    tf.keras.layers.Dense(128, activation='relu'), # Dense(노드의 숫자), 관습적으로 2의 제곱수로 표현
    tf.keras.layers.Dense(1, activation='sigmoid') # output layer는 하나의 노드로 구성(결과값이므로), 결과가 여러개면 layer노드 설정
    # sigmoid는 0~1 사이의 결과값을 return하므로 output layer의 활성함수를 sigmoid로 설정
]) # <- 딥러닝 모델 Sequential을 쓰면 신경망 레이어를 쉽게 만들어줌
```

```python
model.compile(optimizer='adam', loss ='binary_crossentropy', metrics=['accuracy'])
# 손실함수 binary_crossentropy는 0과1사이의 확률 문제에서 많이 사용
```

#### model 학습

```python
model.fit(x, y, epochs = 100) # model.fit(x, y, epochs) epochs는 학습횟수
```

```python
model.fit(np.array(x_train), np.array(y_train), epochs = 1000)
```

#### 모델을 통해 예측

```python
pred = model.predict([[750, 3.70, 3], [400, 2.2, 1]])
```
